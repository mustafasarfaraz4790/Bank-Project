```python
import random
def create():
    account_number = random.randint(10000, 99999)
    account = {
        "Account Number": account_number,
        "Name": input("Enter your name: "),
        "Balance": 0.0,
        "Transactions": []
    }

    try:
        with open("accounts.txt", "a") as file:
            file.write(str(account) + "\n")
        print(f"Account created successfully! Your account number is {account_number}.")
    except Exception as e:
        print(f"An error occurred while creating the account: {e}")




def deposit_money():
    account_number = int(input("Enter your 5-digit account number: "))
    try:
        with open("accounts.txt", "r") as file:
            accounts = file.readlines()
    except FileNotFoundError:
        print("No accounts file found. Please create an account first.")
        return

    updated_accounts = []
    account_found = False

    for i in accounts:
        account = eval(i)
        if account["Account Number"] == account_number:
            account_found = True
            try:
                amount = float(input("Enter the amount to deposit: "))
                if amount <= 0:
                    print("Please enter a positive amount.")
                    return
                account["Balance"] += amount
                account["Transactions"].append(f"Deposit: +{amount}")
                print(f"Successfully deposited {amount}. New balance: {account['Balance']}")
            except ValueError:
                print("Invalid input. Please enter a valid number.")
        updated_accounts.append(account)

    if not account_found:
        print("Account not found.")
        return

    try:
        with open("accounts.txt", "w") as file:  
            for acc in updated_accounts:
                file.write(str(acc) + "\n")
    except Exception as e:
        print(f"An error occurred while saving accounts: {e}")



def withdraw():
    account_number = int(input("Enter your 5-digit account number: "))
    
    try:
        with open("accounts.txt", 'r') as file:
            accounts = file.readlines()  # Read all accounts from the file
    except FileNotFoundError:
        print("No Account Found! Create an account first.")
        return

    updated_accounts = []
    account_found = False

    for line in accounts:
        account = eval(line)  # Convert string to dictionary
        if account["Account Number"] == account_number:
            account_found = True
            amount = float(input("Enter the amount to withdraw: "))

            if amount <= 0:
                print("Please enter a positive value.")
                return
            if amount > account["Balance"]:
                print("Insufficient balance.")
                return

            account["Balance"] -= amount
            account["Transactions"].append(f"Withdraw Amount: -{amount}")
            print(f"Successfully withdrew {amount}. New balance: {account['Balance']}")

        updated_accounts.append(account)

    if not account_found:
        print("Account not found.")
        return

    try:
        with open("accounts.txt", "w") as file:  # Save updated accounts back to the file
            for acc in updated_accounts:
                file.write(str(acc) + "\n")
    except Exception as e:
        print(f"An error occurred while saving accounts: {e}")

    


def print_statement():
    account_number = int(input("Enter your 5-digit account number: "))
    
    try:
        with open("accounts.txt", 'r') as file:
            accounts = file.readlines()
    except FileNotFoundError:
        print("No Account Found! Create an account first.")
        return

    account_found = False

    for line in accounts:
        account = eval(line)
        if account["Account Number"] == account_number:
            account_found = True
            print(f"""
                Transaction Statement
                ----------------------
                Name: {account['Name']}
                Account Number: {account['Account Number']}
                ----------------------
            """)
            if not account["Transactions"]:
                print("No transactions found.")
            else:
                for transaction in account["Transactions"]:
                    print(transaction)
            print(f"Final Balance: {account['Balance']}")
            break

    if not account_found:
        print("Account not found.")


while True:
    print("\nBanking System Menu:")
    print("1. Create Account")
    print("2. Deposit Money")
    print("3. Withdraw")
    print("4. Print Account Details")
    print("5. Exit")

    choice = input("Choose an option: ")
    if choice == "1":
        create()
    elif choice == "2":
        deposit_money()
    elif choice == "3":
        withdraw()
    elif choice == "4":
        print_statement()
    elif choice == "5":
        print("Thank you for using the banking system. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
        
```
